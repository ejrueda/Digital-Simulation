### Picas y Fijas 
Este juego trata de adivinar un número de cuatro cifras el cual no tiene cifras repetidas, Uno de los jugadores piensa un número y el otro lo tiene que adivinar, si el número que dice tiene un dígito igual en la misma posición que el número que hay que adivinar, esto se denomina “fija”, si acierta un dígito pero no en la posición que es, se dice que es “pica”. 

El código a continuación ejecuta dos modos, en un modo el computador piensa un número y el usuario lo tiene que adivinar, y el segundo modo, el usuario piensa un numero y el jugador lo tiene que adivinar.

```python
import numpy as np
import random
import itertools
from time import time 
```

```python
def hallar_picas(y_sub, num_a): #dado el número a adivinar y el número dicho, halla las picas.
    picas=0
    for i in range(4):
        for j in range(4):
            if y_sub[i]==num_a[j]:
                picas=picas+1
    return picas - hallar_fijas(y_sub,num_a)

def validar_num(numero): #para saber si el número es de cuatro cifras diferentes
    return len(np.unique(numero))
    
def hallar_fijas(y_sub, num_a):#dado el número a adivinar y el número dicho, halla las fijas.
    return np.sum(y_sub-num_a==0)

def redim(y): #recibe un entero y lo convierte en un vector
    y_sub=np.zeros(4)
    for i in range(4):
        y_sub[i] = int(y/(10**(3-i)))
        y = y-y_sub[i]*10**(3-i)
    return y_sub
```
